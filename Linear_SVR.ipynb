### Linear Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X=np.array(sample_df['Distance (KM)']).reshape(-1, 1)
Y=np.array(sample_df['Time Minutes']).reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=12)

y_train = y_train.reshape(len(y_train),)
y_test = y_test.reshape(len(y_test),)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt
```

```python
#import sklearn
#sklearn. __version__
```

```python
print(f"Intercept: {lm.intercept_ :,.3f}")
print("Coefficient: {:,.3f}".format(lm.coef_[0]))
    
mae = mean_absolute_error(y_test, lm.predict(X_test))
rmse = sqrt(mean_squared_error(y_test, lm.predict(X_test)))
r_squared = r2_score(y_test, lm.predict(X_test))
print("MAE = {:,.2f} Minutes".format(mae))
print("RMSE = {:,.2f} Minutes".format(rmse))
print("R2 = {:,.2f}%".format(100*r_squared))
```

```python
# Plot outputs
plt.figure(figsize=(14,9))
plt.scatter(x=X, y=Y, edgecolors = "white", linewidths = 0.1, alpha = 0.2)
plt.plot(X_test, lm.predict(X_test), color='red')
plt.xlabel('Distance in KM')
plt.ylabel('Time (Minutes)')
plt.title('Linear Regression Prediction')
plt.show()
```

### SVR
##### Define Performance results function

```python
def svr_results(y_test, X_test, fitted_svr_model):
    print(f"C: {fitted_svr_model.C :,.3f}")
    print(f"Epsilon: {fitted_svr_model.epsilon}")
    print(f"Intercept: {fitted_svr_model.intercept_[0] :,.3f}")
    print(f"Coefficient: {fitted_svr_model.coef_[0] :,.3f}")
    
    mae = mean_absolute_error(y_test, fitted_svr_model.predict(X_test))
    rmse = sqrt(mean_squared_error(y_test, fitted_svr_model.predict(X_test)))
    r_squared = r2_score(y_test, fitted_svr_model.predict(X_test))
    print(f"MAE = {mae :,.2f} Minutes")
    print(f"RMSE = {rmse :,.2f} Minutes")
    print(f"R2 = {r_squared*100 :,.2f}")
    
    perc_within_eps = 100*np.sum(abs(y_test - fitted_svr_model.predict(X_test)) < eps)/len(y_test)
    print(f"Percentage within Epsilon = {perc_within_eps :,.2f} %")
    
    # plot outputs
    plt.figure(figsize=(14,9))
    plt.scatter(x=X, y=Y, edgecolors = "white", linewidths = 0.1, alpha = 0.2,s=50,c="dodgerblue")
    # plt.scatter(X_test,ytest,c="cyan",marker="x",s=50,label="Test Set")
    #plt.scatter(x=scaled_X, y=scaled_Y)
    plt.plot(X_test, fitted_svr_model.predict(X_test), color='red')
    plt.plot(X_test, fitted_svr_model.predict(X_test)+eps, color="black")
    plt.plot(X_test, fitted_svr_model.predict(X_test)-eps, color="black")
    plt.xlabel("Distance in KM", size=18)
    plt.ylabel("Estimated Time of Arrival", size=18)
    plt.title("SVR ETA Prediction", size=30)
    ax = plt.gca() 
    #ax.spines['bottom'].set_color('red')
    #ax.spines['top'].set_color('red')
    #ax.xaxis.label.set_color('red')
    ax.xaxis.label.fontfamily="fantasy"
    #ax.tick_params(axis='x', colors='red')
    #ax.spines['right'].set_visible(False)
    #ax.spines['top'].set_visible(False)
    ax.tick_params(axis = 'both', which = 'major', labelsize = 12)
    f=plt.gcf()
    f.tight_layout()
    plt.show()
```

#### Vary C

```python
test_mae_list = []
perc_within_eps_list = []
test_r2_list = []

eps = 10
c_space = np.linspace(0.01, 10)

for c in c_space:
    varied_svr = LinearSVR(epsilon = eps, C = c, fit_intercept=True, max_iter=10000)
    varied_svr.fit(X_train, y_train)
    
    test_mae = mean_absolute_error(y_test, varied_svr.predict(X_test))
    test_mae_list.append(test_mae)
    
    perc_within_eps = 100*np.sum(abs(y_test-varied_svr.predict(X_test)) <= eps)/len(y_test)
    perc_within_eps_list.append(perc_within_eps)
    
    test_r2 = r2_score(y_test, varied_svr.predict(X_test))
    test_r2_list.append(test_r2)
```

```python
#print(len(test_mae_list))
#print(len(perc_within_eps_list))
#print(len(c_space))
print(f"Minimum MAE: {min(test_mae_list)} \nMax MAE {max(test_mae_list)}")
print(f"Minimum % Within Epsilon: {min(perc_within_eps_list)} \n Max % Within Eps: {max(perc_within_eps_list)}")
```

```python
fig, ax1 = plt.subplots(figsize=(12,7))

color = "green"
ax1.set_xlabel('C')
ax1.set_ylabel("% within epsilon", color = color)
ax1.scatter(c_space, perc_within_eps_list, color=color)
ax1.tick_params(axis='y', labelcolor=color)

color = "blue"
ax2 = ax1.twinx() # Instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Test MAE', color = color)
ax2.scatter(c_space, test_mae_list, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.show()
```

```python
m = max(perc_within_eps_list)
inds = [i for i, j in enumerate(perc_within_eps_list) if j == m]
C = c_space[inds[0]]

print("best C =", C)
```

Best C From GridSearch

```python
eps = 10
svr_best_C = LinearSVR(epsilon = eps, C = C, fit_intercept=True)
svr_best_C.fit(X_train, y_train)
svr_results(y_test, X_test, svr_best_C)
```

#### Grid Search over C and Epsilon

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
```

```python
grid = {
    'C': np.linspace(0.01, 10),
    'epsilon':np.linspace(1, 10)
}

svr_gridsearch = LinearSVR(fit_intercept=True, max_iter=10000)
```

#### Scoring Criteria = MAE

```python
grid_svr = GridSearchCV(svr_gridsearch, grid, scoring='neg_mean_absolute_error', cv=5)
```

```python
grid_svr.fit(X_train, y_train)
```

```python
best_grid_svr_mae = grid_svr.best_estimator_
best_grid_svr_mae.fit(X_train, y_train)
```

```python
svr_results(y_test, X_test, best_grid_svr_mae)
```

#### Scoring Criteria = % within Epsilon

```python
grid = {
    'C':np.linspace(0.01, 10),
    'epsilon': np.linspace(1 , 10)
}

svr_gridsearch = LinearSVR(fit_intercept=True, max_iter=10000)
```
