```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as  np
```

```python
df = pd.read_csv('Sendy/Train.csv')
rider_df = pd.read_csv('Sendy/Riders.csv')
```

```python
df.head()
```

```python
rider_df.head()
```

```python
merged = df.merge(rider_df, on='Rider Id')
```

```python
merged['Time Minutes'] = merged['Time from Pickup to Arrival']/60
merged.head()
```

```python
abs(merged.corr())['Time from Pickup to Arrival'].sort_values(ascending=False)
```

```python
subset_df = merged[['Distance (KM)', 'Destination Long','Destination Lat','Pickup Long','Pickup Lat','Average_Rating','No_Of_Orders','Time Minutes']]
subset_df.head()
```

```python
subset_df.describe()
```

```python
### shuffle the dataset to allow for subsetting
subset_df = subset_df.sample(frac=1).reset_index(drop=True)
```

```python
### Filter data to remain only with filter(time_from_pickup_to_arrival >= 600 & distance_km >= 5 & distance_km <=40) %>% head(1000)
subset_df = subset_df.loc[(subset_df['Time Minutes'] >= 5) & ((subset_df['Distance (KM)'] >= 2) & (subset_df['Distance (KM)'] <= 35))]
```

```python
sample_df = subset_df.iloc[0:15000,]
```

```python
X = sample_df.iloc[:, 0:-1]
y = sample_df.iloc[:, -1]
```

```python
y.head()
```

```python
X.head()
```

### Split the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=15, random_state=0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = sqrt(mean_squared_error(y_test, y_pred))
```

```python
print(f'{mae} minutes')
```
